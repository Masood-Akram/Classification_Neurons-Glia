```python
#Importing the libraries
import pandas as pd
import numpy as np
import itertools
import random
```

```python
#Loading a dataset and checking the dataset shape
df = pd.read_excel('Neuron_Data_Final.xlsx', header=None)
df.shape
```

```python
#Converting the dataset into a list
#Checking for missing values
dfint = df.values.tolist()

nan = np.nan
df1 = [[x for x in y if not np.isnan(x)] for y in dfint]
```

```python
#Importing the random sampling library
#Making a function for random combinations without repleacement
from random import seed, sample

seed(10)

def random_combinations(seq, size, num):
    combos = set()
    while len(combos) < num:
        item = sample(seq, size)
        combos.add(frozenset(item))
    return [tuple(u) for u in combos]
```

```python
#Taking hundred random samples of 1 branch value from each Neuron
seed(10)

l_1 = []
for i in range(100):

    
    for num in df1:
    # print(num)
        len_of_list = len(num)
        
        if len_of_list > 0:
            combos = random_combinations(num, 1, 1)
        elif len_of_list < 1:
            combos = random_combinations(num, 0, 1)
        l_1.append(combos)
    
    
List1 = l_1

#Taking hundred random samples of 2 branch value from each Neuron
l_2 = []

for i in range(100):

    
    for num in df1:
    # print(num)
        len_of_list = len(num)
        
        if len_of_list > 1:
            combos = random_combinations(num, 2, 1)
        elif len_of_list < 2:
            combos = random_combinations(num, len_of_list, 1)
        l_2.append(combos)
    
    

List2 = l_2

#Taking hundred random samples of 3 branch value from each Neuron
l_3 = []

for i in range(100):

    
    for num in df1:
    # print(num)
        len_of_list = len(num)
        
        if len_of_list > 2:
            combos = random_combinations(num, 3, 1)
        elif len_of_list < 3:
            combos = random_combinations(num, len_of_list, 1)
        l_3.append(combos)
    
    

List3 = l_3

#Taking hundred random samples of 4 branch value from each Neuron
l_4 = []

for i in range(100):

    
    for num in df1:
    # print(num)
        len_of_list = len(num)
        
        if len_of_list > 3:
            combos = random_combinations(num, 4, 1)
        elif len_of_list < 4:
            combos = random_combinations(num, len_of_list, 1)
        l_4.append(combos)
    
    

List4 = l_4

#Taking hundred random samples of 5 branch value from each Neuron
l_5 = []

for i in range(100):

    
    for num in df1:
    # print(num)
        len_of_list = len(num)
        
        if len_of_list > 4:
            combos = random_combinations(num, 5, 1)
        elif len_of_list < 5:
            combos = random_combinations(num, len_of_list, 1)
        l_5.append(combos)
    
    

List5 = l_5

#Taking hundred random samples of 10 branch value from each Neuron
l_10 = []

for i in range(100):

    
    for num in df1:
    # print(num)
        len_of_list = len(num)
        
        if len_of_list > 9:
            combos = random_combinations(num, 10, 1)
        elif len_of_list < 10:
            combos = random_combinations(num, len_of_list, 1)
        l_10.append(combos)
    
    

List10 = l_10

#Taking hundred random samples of 15 branch value from each Neuron
l_15 = []

for i in range(100):

    
    for num in df1:
    # print(num)
        len_of_list = len(num)
        
        if len_of_list > 14:
            combos = random_combinations(num, 15, 1)
        elif len_of_list < 15:
            combos = random_combinations(num, len_of_list, 1)
        l_15.append(combos)
    
    

List15 = l_15
```

```python
#Importing the statistic library
from statistics import mean

#Taking the average of 100 samples of each Neuron for 1 branch
avg1 = [[mean(sub_list) for sub_list in list] for list in List1]

#Taking the average of 100 samples of each Neuron for 2 branches 
avg2 = [[mean(sub_list) for sub_list in list] for list in List2]

#Taking the average of 100 samples of each Neuron for 3 branches 
avg3 = [[mean(sub_list) for sub_list in list] for list in List3]

#Taking the average of 100 samples of each Neuron for 4 branches 
avg4 = [[mean(sub_list) for sub_list in list] for list in List4]

#Taking the average of 100 samples of each Neuron for 5 branches 
avg5 = [[mean(sub_list) for sub_list in list] for list in List5]

#Taking the average of 100 samples of each Neuron for 10 branches 
avg10 = [[mean(sub_list) for sub_list in list] for list in List10]

#Taking the average of 100 samples of each Neuron for 15 branches 
avg15 = [[mean(sub_list) for sub_list in list] for list in List15]
```

```python
#Converting the lists to arrays for 1 branch
A1 = np.array(avg1)

#Converting the lists to arrays for 2 branches
A2 = np.array(avg2)

#Converting the lists to arrays for 3 branches
A3 = np.array(avg3)

#Converting the lists to arrays for 4 branches
A4 = np.array(avg4)

#Converting the lists to arrays for 5 branches
A5 = np.array(avg5)

#Converting the lists to arrays for 10 branches
A10 = np.array(avg10)

#Converting the lists to arrays for 15 branches
A15 = np.array(avg15)
```
